The plt.plot() function in Matplotlib is versatile and allows for many optional parameters to control every aspect of the plot, from colors to line styles and transparency. Below is a breakdown of plt.plot() parameters and how they work

-----

`plt.plot(x, y, <optional parameters>)`


Parameters

x (array-like, optional):

The data for the x-axis values. If only y is provided, x is assumed to be the indices of y.

y (array-like, optional):

The data for the y-axis values.
color or c (str or array-like, optional):

Sets the color of the line. Can use color names (e.g., 'blue'), hex color codes (e.g., '#FFDD44'), or single-character color abbreviations (e.g., 'b' for blue, 'r' for red).
marker (str, optional):

Defines the marker style at each data point (e.g., 'o' for circles, 'x' for crosses, '^' for triangles).
linestyle or ls (str, optional):

Sets the line style. Examples:
'-' for a solid line,
'--' for a dashed line,
'-.' for dash-dot,
':' for a dotted line.
linewidth or lw (float, optional):

Defines the width of the line. Higher values make the line thicker.
label (str, optional):

Label for the plot, which can be used in legends. This label will appear in a legend if plt.legend() is used.
alpha (float, optional):

Controls the transparency of the line, with values from 0 (fully transparent) to 1 (fully opaque).
markerfacecolor or mfc (str or array-like, optional):

Sets the color of the marker face (interior). Accepts color names, hex codes, or abbreviations.
markeredgecolor or mec (str or array-like, optional):

Sets the color of the marker edge (outline).
markersize or ms (float, optional):

Sets the size of the markers.
markevery (int, float, or sequence, optional):

Specifies whether to plot markers at every data point or only at some. Useful for large datasets.
zorder (int, optional):

Sets the order of layering for multiple plots, where higher zorder values are plotted on top.
drawstyle (str, optional):

Determines how line segments are drawn. Options include 'default', 'steps-pre', 'steps-mid', and 'steps-post'.
dash_capstyle (str, optional):

Defines the cap style for dashes. Options: 'butt', 'round', 'projecting'.
dash_joinstyle (str, optional):

Sets the join style for dashes. Options: 'miter', 'round', 'bevel'.
solid_capstyle (str, optional):

Sets the cap style for solid lines. Options: 'butt', 'round', 'projecting'.
solid_joinstyle (str, optional):

Defines the join style for solid lines. Options: 'miter', 'round', 'bevel'.
antialiased or aa (bool, optional):

Toggles antialiasing, which can make lines look smoother.
scalex and scaley (bool, optional):

If True, auto-scales the respective axis based on the plot's data.
data (dict or data object, optional):

Passes data from a dictionary or similar data structure to x and y






```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z=[12,13,14,15,16]
```

### 1. Line Graph

```python
plt.plot(x, y,marker='o')
plt.xlabel("X Axis Label")
plt.ylabel("Y Axis Label")
plt.title("Line Plot Example")
plt.show()
```

###2. Bar Graph

`plt.bar(x, height, width=0.8, bottom=None, align='center', **kwargs)`

Parameters
Here are the primary parameters for plt.bar():

1. x: (array-like)

   The x-coordinates of the bars.

2. height: (array-like)

  The height of the bars, usually representing the values you want to plot.

3. width: (float or array-like, optional)

  The width of each bar. Default is 0.8.
4. bottom: (array-like, optional)

  The y-coordinate of the bottom of each bar. Default is 0, which places the bar on the x-axis.
5. align: ({'center', 'edge'}, optional)

  Alignment of the bars to the x coordinates. Default is 'center'.

6. color, edgecolor, linewidth, etc. (optional)
   Additional styling parameters, like bar color, edge color, and line width.

```python
plt.bar(x, y,width=0.8,color='red')
plt.xlabel("X Axis Label")
plt.ylabel("Y Axis Label")
plt.title("Line Plot Example")
plt.show()
```

###3. Scatter Plot

`plt.scatter(x, y, s=None, c=None, marker=None, cmap=None, **kwargs)`



Parameters
1. x: (array-like)

  The x-coordinates of the points.
2. y: (array-like)

  The y-coordinates of the points.
3. s: (scalar or array-like, optional)

  Size of each point (defaults to 20). You can pass an array to set different sizes for each point.
4. c: (color, sequence, or array-like, optional)

  Color of the points. This can be a single color format string or an array of colors for each point.
5. marker: (string, optional)

  The shape of each point (e.g., 'o' for circles, 's' for squares, '^' for triangles). Default is 'o'.
6. cmap: (Colormap, optional)

  Colormap for mapping numerical data to colors if c is an array of values.
7. alpha: (float, optional)

  Transparency level of points (value between 0 and 1, where 0 is fully transparent).
8. edgecolor: (color or array-like, optional)

  Color of the edges around each point.

```python
plt.scatter(x, y, alpha=0.5)
plt.colorbar()
plt.title("Scatter Plot Example")
plt.show()
```

### 4. Histogram

A histogram is a type of bar plot that shows the distribution of a dataset by dividing it into bins and counting the frequency of values that fall into each bin. In Matplotlib, you can create a histogram using the plt.hist() function.

`plt.hist(x, bins=None, range=None, density=False, cumulative=False, **kwargs)`

Parameters
1. x: (array-like)

  The data you want to plot.
2. bins: (int or sequence, optional)

  The number of bins (bars) in the histogram. If an integer, it defines the number of equal-width bins. If a sequence, it defines the specific bin edges.
3. range: (tuple or None, optional)

  The lower and upper range of bins. If not provided, range is (x.min(), x.max()).
4. density: (bool, optional)

  If True, it normalizes the histogram so the area under the histogram equals 1 (useful for probability density).
5. cumulative: (bool, optional)

  If True, the histogram is cumulative, meaning each bin sums all frequencies up to that bin.
6. color, edgecolor, etc. (optional)

  Additional styling options, such as bar color and edge color

```python
plt.hist(x, bins=30, color='purple', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```
