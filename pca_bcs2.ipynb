```python
import pandas as pd
```

```python
df=pd.read_csv('/content/breast_cancer_survival.csv')
```

```python
df
```

```python
df.columns[6:11]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in df.columns:
  print(i)
```

```python
for each in df.columns[6:11]:
  df[each]=le.fit_transform(df[each])
```

```python
df
```

```python
df['Surgery_type']=le.fit_transform(df['Surgery_type'])
```

```python
df
```

```python
df['Gender']=le.fit_transform(df['Gender'])
```

```python
df
```

```python
df['Patient_Status']=le.fit_transform(df['Patient_Status'])
```

```python
df.drop(['Date_of_Last_Visit'],axis=1,inplace=True)
```

```python
df.drop(['Date_of_Surgery'],axis=1,inplace=True)
```

```python
df
```

```python
y=df['Patient_Status']
y
```

```python
x=df.drop(['Patient_Status'],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
ts=[0.2,0.3,0.4]
```

```python
model=KNeighborsClassifier(n_neighbors = 5)
```

```python
for each in ts:
  X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=each,random_state=1)
  print(X_train.shape)
  print(X_test.shape)
  print(y_train.shape)
  print(y_test.shape)
  model.fit(X_train,y_train)
  y_pred = model.predict(X_test)
  acc=accuracy_score(y_test,y_pred)
  print(acc)
```

```python
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
```

```python
pca=PCA(n_components=2)
```

```python
x_train_reduced=pca.fit_transform(X_train)
x_test_reduced=pca.transform(X_test)
```

```python
model_reduced=DecisionTreeClassifier()
```

```python
model_reduced.fit(x_train_reduced,y_train)
```

```python
y_reduced_pred=model_reduced.predict(x_test_reduced)
```

```python
accuracy_with_reduction = accuracy_score(y_test, y_reduced_pred)
```

```python
accuracy_with_reduction
```

```python
print("Accuracy without dimensionality reduction:", acc)
print("Accuracy with dimensionality reduction (PCA):", accuracy_with_reduction)
```
