The expression `df.isnull().sum()` is used to count the number of missing (null) values in each column of a pandas DataFrame.

Breakdown:
df.isnull(): This returns a DataFrame of the same shape as df, where each element is True if it is NaN (or None), and False otherwise.

.sum(): When applied to a DataFrame of Boolean values, True is treated as 1 and False as 0. So, .sum() gives the count of True values (i.e., missing values) for each column

The `df.dropna()` method in pandas is used to remove rows or columns that contain missing values (NaN or None) from a DataFrame.

Key Features:
Removes missing values: It drops rows or columns that have any missing data (NaN or None).
Flexible options: You can choose to drop rows or columns based on specific conditions (e.g., drop rows with any NaN values or only those with all NaN values).

Parameters:

1. axis:

  0 or 'index' (default) to drop rows.
  1 or 'columns' to drop columns.


2. how:

    'any' (default): Drops a row/column if any of its values are
     missing.
    
    'all': Drops a row/column only if all of its values are missing.

3. thresh: Keeps rows/columns that have at least a certain number of
   non-NaN values. For example, thresh=2 keeps rows that have at least 2 non-NaN values.

4. subset: Allows specifying which rows/columns to check for missing
   values. This is useful when you want to drop based on a specific set of columns.

5. inplace: If True, the operation is done in place and the DataFrame is modified. If False (default), it returns a new DataFrame without modifying the original.

```python
#df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

The `df.fillna()` method in pandas takes up to 6 parameters, but typically you only use the first two in most common cases


Parameters of fillna():
1. value:

This can be a scalar, dictionary, Series, or DataFrame.
It is the value to replace NaN with. In your example, df.mean() is used, which fills missing values with the mean of each column.

2. method:

Specifies the method used for filling missing values.

  Options: 'ffill' (forward fill) or 'bfill' (backward fill).
  If provided, it ignores the value parameter.


3. axis:

Specifies whether to fill along rows (0) or columns (1).
Default is 0 (fills along rows).

4. inplace:

If True, modifies the original DataFrame.
If False (default), returns a new DataFrame with the missing values filled.

5. limit:

Limits the number of missing values to fill.
For example, limit=1 will only fill the first missing value encountered in each column.

6. downcast:

Used to downcast the data type if possible.
For example, if filling with integers, it tries to downcast to the smallest possible integer type.

The **df.duplicated()** method in pandas is used to identify duplicate rows in a DataFrame. It returns a Boolean Series where True represents a row that is a duplicate of a previous row, and False indicates a unique row.

`df.duplicated(subset=None, keep='first')`

Parameters:

1. subset (optional):

Specifies which columns to consider when identifying duplicates.
If None (default), it checks for duplicates across all columns.
You can pass a single column label or a list of column labels to focus on certain columns.

2. keep (optional, default 'first'):

Determines which duplicate rows to mark as True:
'first' (default): Marks all duplicates after the first occurrence as True.
'last': Marks all duplicates before the last occurrence as True.
False: Marks all duplicate rows as True.

Returns:
A Boolean Series where:
True indicates a duplicate row.
False indicates a unique row.


`df.duplicated(subset=['A', 'B'])` **#Checks duplicates based only on columns A and B**


`df.duplicated(keep='last')`  **# Marks earlier duplicates instead of later ones**

`df.duplicated(keep=False)`  







```python
df.duplicated()
```

```python
df.duplicated(subset=['year'])
```

The expression` df.duplicated().sum()` is used to count the number of duplicate rows in a pandas DataFrame.

Explanation:
df.duplicated():

This method returns a Boolean Series, where True represents a duplicate row, and False represents a unique row.
.sum():

Since True is treated as 1 and False is treated as 0 in pandas, applying .sum() on the Boolean Series will count the number of True values, which corresponds to the number of duplicate rows.

```python
df.duplicated().sum()
```

The `df.drop_duplicates()` method in pandas is used to remove duplicate rows from a DataFrame. It can modify the DataFrame in place or return a new one, depending on the parameters specified.

df.drop_duplicates(subset=None, keep='first', inplace=False)

**Parameters:**

subset (optional):

Specifies which columns to consider when identifying duplicates.
Default: None (checks for duplicates across all columns).
Can be a single column label or a list of column labels to check for duplicates only in specific columns.

Example
`df.drop_duplicates(subset=['A', 'B'])`

keep (optional):

Determines which duplicates to keep:
'first' (default): Keep the first occurrence of each duplicate and remove subsequent ones.
'last': Keep the last occurrence of each duplicate and remove earlier ones.
False: Remove all occurrences of duplicates.

Example:

`df.drop_duplicates(keep='last')`  **# Keeps the last occurrence of each duplicate**

`df.drop_duplicates(keep=False) `   **# Removes all occurrences of duplicates**


```python
df.drop_duplicates(subset=['name'])
```
