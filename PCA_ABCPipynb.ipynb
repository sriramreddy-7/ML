```python
import pandas as pd
```

```python
df=pd.read_csv('/content/breast_cancer_survival.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df.drop(['Date_of_Surgery','Date_of_Last_Visit'],inplace=True,axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.dtypes
```

```python
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=le.fit_transform(df[i])
```

```python
df
```

```python
y=df['Patient_Status']
y
```

```python
x=df.drop('Patient_Status',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
model1=SVC(kernel='linear',random_state=1)
model2=KNeighborsClassifier(n_neighbors=15)
model3=LogisticRegression(max_iter=1000,random_state=42)
```

```python
model1.fit(X_train,Y_train)
model2.fit(X_train,Y_train)
model3.fit(X_train,Y_train)
```

```python
y_pred1=model1.predict(X_test)
y_pred2=model2.predict(X_test)
y_pred3=model3.predict(X_test)
```

```python
accuracy1=accuracy_score(Y_test,y_pred1)
accuracy2=accuracy_score(Y_test,y_pred2)
accuracy3=accuracy_score(Y_test,y_pred3)
```

```python
print(accuracy1*100)
print(accuracy2*100)
print(accuracy3*100)
```

```python
cf1=classification_report(Y_test,y_pred1)
print(cf1)
cf2=classification_report(Y_test,y_pred2)
print(cf2)
cf3=classification_report(Y_test,y_pred3)
print(cf3)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=7)
```

```python
X_train_pca=pca.fit_transform(X_train)
X_test_pca=pca.transform(X_test)
```

```python
X_train_pca
X_test_pca
```

```python
# Initialize classifiers again for re-training
svm_pca = SVC(kernel='linear', random_state=42)
knn_pca = KNeighborsClassifier(n_neighbors=5)
log_reg_pca = LogisticRegression(random_state=42, max_iter=1000)

# Train and test the SVM model on PCA-transformed data
svm_pca.fit(X_train_pca, Y_train)
y_pred_svm_pca = svm_pca.predict(X_test_pca)
print("SVM with PCA Accuracy:", accuracy_score(Y_test, y_pred_svm_pca))
print("SVM with PCA Classification Report:\n", classification_report(Y_test, y_pred_svm_pca))

# Train and test the KNN model on PCA-transformed data
knn_pca.fit(X_train_pca, Y_train)
y_pred_knn_pca = knn_pca.predict(X_test_pca)
print("KNN with PCA Accuracy:", accuracy_score(Y_test, y_pred_knn_pca))
print("KNN with PCA Classification Report:\n", classification_report(Y_test, y_pred_knn_pca))

# Train and test the Logistic Regression model on PCA-transformed data
log_reg_pca.fit(X_train_pca, Y_train)
y_pred_log_reg_pca = log_reg_pca.predict(X_test_pca)
print("Logistic Regression with PCA Accuracy:", accuracy_score(Y_test, y_pred_log_reg_pca))
print("Logistic Regression with PCA Classification Report:\n", classification_report(Y_test, y_pred_log_reg_pca))
```

```python
# Import PCA
from sklearn.decomposition import PCA

# Set the number of principal components (you can adjust this based on your dataset)
n_components = 5  # Reducing to 5 components for illustration

# Apply PCA to the training and test sets
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

y_test=Y_test
y_train=Y_train
# Initialize classifiers again for re-training
svm_pca = SVC(kernel='linear', random_state=42)
knn_pca = KNeighborsClassifier(n_neighbors=5)
log_reg_pca = LogisticRegression(random_state=42, max_iter=1000)

# Train and test the SVM model on PCA-transformed data
svm_pca.fit(X_train_pca, y_train)
y_pred_svm_pca = svm_pca.predict(X_test_pca)
print("SVM with PCA Accuracy:", accuracy_score(y_test, y_pred_svm_pca))
print("SVM with PCA Classification Report:\n", classification_report(y_test, y_pred_svm_pca))

# Train and test the KNN model on PCA-transformed data
knn_pca.fit(X_train_pca, y_train)
y_pred_knn_pca = knn_pca.predict(X_test_pca)
print("KNN with PCA Accuracy:", accuracy_score(y_test, y_pred_knn_pca))
print("KNN with PCA Classification Report:\n", classification_report(y_test, y_pred_knn_pca))

# Train and test the Logistic Regression model on PCA-transformed data
log_reg_pca.fit(X_train_pca, y_train)
y_pred_log_reg_pca = log_reg_pca.predict(X_test_pca)
print("Logistic Regression with PCA Accuracy:", accuracy_score(y_test, y_pred_log_reg_pca))
print("Logistic Regression with PCA Classification Report:\n", classification_report(y_test, y_pred_log_reg_pca))
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

best_accuracy = 0
best_n_components = 0

for n in range(1, X_train.shape[1] + 1):
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # Train model on PCA-transformed data
    model = SVC()  # or KNN, Logistic Regression, etc.
    model.fit(X_train_pca, Y_train)
    y_pred = model.predict(X_test_pca)
    accuracy = accuracy_score(Y_test, y_pred)

    print(f"Number of PCA Components: {n}, Accuracy: {accuracy}")

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_n_components = n

print(f"Best Number of PCA Components: {best_n_components}, Best Accuracy: {best_accuracy}")
```
